## **Historia, ciencia y programación: la “pluma de agua”**

> **Relato de partida**  
> **¿Por qué los costeños le decimos “pluma” a la llave del agua?**  
>  
> En el Caribe colombiano, abrir la “pluma” no es solo una acción cotidiana: es una expresión cargada de historia, técnica y cultura. Aunque algunos se sorprendan o incluso se burlen del término, lo cierto es que decir “pluma” en vez de “llave” no es un error, sino una herencia viva que conecta nuestro lenguaje con siglos de evolución hidráulica.  
>  
> Mucho antes de que existieran los grifos modernos, el agua se distribuía mediante sistemas regulados con precisión. En España, por ejemplo, se usaba la **“pluma de agua”** como unidad de medida para calcular el caudal que se entregaba a cada hogar. Aproximadamente, una pluma equivalía a **0.025 litros por segundo**. En algunos lugares, ese flujo se regulaba mediante orificios calibrados y dispositivos llamados **plumeros**, que permitían controlar la salida del agua.  
>  
> Durante la época colonial, muchas de estas prácticas técnicas cruzaron el Atlántico y llegaron a ciudades como Cartagena, Barranquilla y Santa Marta. Con el tiempo, el término “pluma” dejó de ser solamente técnico y pasó al habla cotidiana como sinónimo de la llave del agua.  
>  
> Así, una palabra del uso diario conserva una memoria histórica, tecnológica y cultural.

## 2. Situación problema

Un hogar recibe una cierta cantidad de **plumas de agua**.  
Se sabe que:

- **1 pluma = 0.025 litros por segundo**

Con esta información, el estudiante debe construir un programa que permita calcular:

1. el caudal total en litros por segundo;
2. el consumo diario en litros;
3. el consumo mensual en litros (suponiendo 30 días);
4. una clasificación del consumo:
   - **Bajo**: menos de 1000 litros por día
   - **Medio**: entre 1000 y 3000 litros por día
   - **Alto**: más de 3000 litros por día

## 5. Algoritmo propuesto

### Algoritmo en lenguaje natural

1. Solicitar al usuario el número de plumas de agua.
2. Definir el valor de una pluma en litros por segundo.
3. Definir la cantidad de segundos que tiene un día.
4. Calcular el caudal total.
5. Calcular el consumo diario.
6. Calcular el consumo mensual.
7. Determinar si el consumo es bajo, medio o alto.

````python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas
# Escriba aquí su código
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema
# Escriba aquí su código
Litros_por_pluma = 0.025
segundos_por_dia = 86400

# 3. Calcular el caudal total
# Escriba aquí su código
Caudal_total = (plumas * Litros_por_pluma)

# 4. Calcular el consumo diario
# Escriba aquí su código
Consumo_diario = (Caudal_total * segundos_por_dia)

# 5. Calcular el consumo mensual
# Escriba aquí su código
Consumo_mensual = (Consumo_diario * 30)

# 6. Clasificar el consumo
# Escriba aquí su código
if Consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif Consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 7. Mostrar resultados
# Escriba aquí su código
print("Caudal total:", Caudal_total, "litros por segundo")
print("Consumo diario:", Consumo_diario, "litros")
print("Consumo mensual:", Consumo_mensual, "litros")
print("Nivel de consumo:", nivel_consumo)
````

## Documentación Técnica: Modelo de Consumo "Pluma de Agua"

### 1. Propósito General del Programa

El objetivo de este software es automatizar el cálculo del consumo de agua potable en una vivienda a partir de la unidad tradicional denominada **"pluma"**. El programa normaliza esta unidad al Sistema Internacional ($L/s$), proyecta el consumo a escalas diarias y mensuales, y proporciona un diagnóstico cualitativo del uso del recurso hídrico.

### 2. Explicación de Entradas, Procesos y Salidas

* **Entradas:** El programa requiere un dato numérico (punto flotante) que represente la cantidad de plumas asignadas al hogar.
* **Procesos:**
1. **Conversión de Caudal:** Se transforma el número de plumas a litros por segundo utilizando la constante $0.025$.
2. **Cálculo Temporal:** Se escala el caudal a un volumen diario (multiplicando por los segundos de un día) y mensual (proyectando a 30 días).
3. **Clasificación Lógica:** Se evalúa el volumen diario mediante umbrales definidos para categorizar el consumo.


* **Salidas:** Un reporte impreso que muestra el caudal total ($L/s$), el consumo diario ($L$), el consumo mensual ($L$) y el nivel de consumo (Bajo, Medio o Alto).

### 3. Descripción de Variables Importantes

| Variable | Tipo | Descripción |
| --- | --- | --- |
| `plumas` | `float` | Cantidad de plumas ingresadas por el usuario. |
| `LITROS_POR_PLUMA` | `float` | Constante que define el caudal de una pluma ($0.025$ $L/s$). |
| `SEGUNDOS_DIA` | `int` | Constante de tiempo ($86400$ segundos). |
| `caudal_total` | `float` | Resultado de la operación: $plumas \cdot 0.025$. |
| `consumo_diario` | `float` | Volumen total de agua en 24 horas. |
| `nivel` | `string` | Etiqueta cualitativa resultante de la estructura condicional. |

### 4. Explicación de las Estructuras Usadas

1. **Estructura Secuencial:** El programa ejecuta los cálculos de forma lineal, donde cada resultado depende del paso anterior (Caudal $\rightarrow$ Diario $\rightarrow$ Mensual).
2. **Estructura Selectiva (`if-elif-else`):** Se emplea una lógica de decisión múltiple. Es la estructura ideal aquí porque los rangos de consumo son **mutuamente excluyentes** (un consumo no puede ser bajo y alto al mismo tiempo).
3. **Manejo de Excepciones (`try-except`):** (Opcional pero recomendado) Se utiliza para evitar que el programa falle si el usuario ingresa texto en lugar de números.

### 5. Observaciones sobre Claridad, Estilo y Mejoras

# Función principal para ejecutar el modelo hídrico
def modelar_pluma_agua():
    # --- 1. Entrada de datos ---
    # Solicitamos el dato al usuario y lo convertimos a decimal (float)
    plumas = float(input("Ingrese el número de plumas asignadas: "))

    # --- 2. Definición de constantes del problema ---
    # Definimos valores fijos para evitar "números mágicos" en el código
    LITROS_POR_PLUMA = 0.025  # Equivalencia técnica
    SEGUNDOS_DIA = 86400      # 24 horas * 3600 segundos

    # --- 3, 4 y 5. Procesamiento y Cálculos ---
    # Calculamos el caudal instantáneo
    caudal_total = plumas * LITROS_POR_PLUMA
    
    # Calculamos el volumen diario mediante la relación de tiempo
    consumo_diario = caudal_total * SEGUNDOS_DIA
    
    # Proyectamos el consumo a un mes estándar de 30 días
    consumo_mensual = consumo_diario * 30

    # --- 6. Clasificación del consumo ---
    # Evaluamos el consumo diario para asignar una categoría
    if consumo_diario < 1000:
        nivel = "Bajo"
    elif 1000 <= consumo_diario <= 3000:
        nivel = "Medio"
    else:
        nivel = "Alto"

    # --- 7. Mostrar resultados ---
    # Imprimimos los datos procesados con formato claro
    print("\n--- REPORTE TÉCNICO DE CONSUMO ---")
    print(f"Caudal Total: {caudal_total:.3f} L/s")
    print(f"Consumo Diario: {consumo_diario:.2f} Litros")
    print(f"Consumo Mensual: {consumo_mensual:.2f} Litros")
    print(f"Clasificación: {nivel}")

# Punto de entrada del script
if __name__ == "__main__":
    modelar_pluma_agua()
