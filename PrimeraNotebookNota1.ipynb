# 1. Problema de ingeniería (ejemplo)

Un ingeniero agrícola necesita calcular el **área de un terreno rectangular** para estimar cuántos cultivos puede sembrar.

La fórmula es:

[
Área = largo \times ancho
]

---

# 2. Definición del algoritmo

Primero definimos **Entradas – Proceso – Salidas**.

### Entradas

Son los datos que el programa necesita.

* `largo` del terreno (m)
* `ancho` del terreno (m)

### Proceso

Operación que realiza el algoritmo.

* Calcular el área multiplicando largo por ancho.

### Salida

Resultado que el programa mostrará.

* `area` del terreno en **m²**

---

# 3. Representación del algoritmo

### Algoritmo (idea lógica)

1. Leer el largo del terreno.
2. Leer el ancho del terreno.
3. Calcular el área.
4. Mostrar el resultado.

---

# 4. Pseudocódigo

Antes de programar, en ingeniería usamos pseudocódigo.

```
INICIO

Leer largo
Leer ancho

area = largo * ancho

Mostrar area

FIN
```

---

# 5. Código en Python

```python
# Programa para calcular el área de un terreno rectangular
# Autor: Estudiante de Ingeniería

# Entrada de datos
largo = float(input("Ingrese el largo del terreno en metros: "))
ancho = float(input("Ingrese el ancho del terreno en metros: "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "m²")
```

---

# 6. Ejemplo de ejecución

Entrada:

```
largo = 50
ancho = 20
```

Cálculo:

```
area = 50 * 20 = 1000
```

Salida:

```
El área del terreno es: 1000 m²
```

### Tu primer reto algorítmico

Antes de darte el código "masticado", quiero que pensemos como ingenieros. Imagina que necesitas calcular el **volumen de agua necesario para un tanque de almacenamiento cilíndrico** en una finca.

Para resolver esto, necesitamos:

1. Definir el radio y la altura del tanque.
2. Aplicar la fórmula matemática: $V = \pi \cdot r^2 \cdot h$.
3. Mostrar el resultado de forma clara.

Aquí tienes un ejemplo de cómo se vería esto en Python, aplicando **buenas prácticas** (nombres de variables descriptivos):

````python
import math

# Definición de parámetros del tanque (en metros)
radio_m = 2.5
altura_m = 4.0

# Cálculo del volumen usando la librería math para obtener el valor de pi
volumen_m3 = math.pi * (radio_m ** 2) * altura_m

# Mostramos el resultado con formato
print(f"El volumen del tanque es: {volumen_m3:.2f} metros cúbicos.")
````

### ¿Cómo lo harías más útil?

Ese código funciona, pero es "estático". Si mañana el tanque cambia de tamaño, tendrías que editar el código manualmente. En ingeniería, buscamos que nuestras herramientas sean **reutilizables**.
