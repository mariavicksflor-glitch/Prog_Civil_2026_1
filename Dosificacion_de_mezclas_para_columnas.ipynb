**Dosificacion de mezclas**

**Planteamiento**:
Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular. Se sabe que para un concreto estándar de 210 kg/cm2
 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos**:

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$v = π * r^2 * h$$
V=π×r2×h
. (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos: Sacos=Volumen×9.5
.
$$Sacos = v * 9.5$$
**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58
 (el estudiante verá aquí la importancia de los tipos decimales).

```python
# ENTRADA: Solicitar la altura de la columna en metros.
# Solicitar el diametro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: ").replace(',', '.'))

# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula los sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# SALIDA: Imprimir el volumen y la cantidad de sacos.
print(f"El volumen de la columna es: {v:.2f} m3")
print(f"La cantidad de sacos necesarios es: {sacos:.2f}")
```

A continuación, se detalla el código utilizado para calcular el volumen de una columna circular y la cantidad de sacos de cemento necesarios:

*   `# ENTRADA: Solicitar la altura de la columna en metros.` y `# Solicitar el diametro de la columna en metros.`: Estas son líneas de comentario que indican la sección del código dedicada a la entrada de datos por parte del usuario.

*   `h = float(input("Ingrese la altura de la columna en metros: "))`: Esta línea solicita al usuario que ingrese la altura de la columna en metros. La función `input()` lee la entrada como una cadena de texto, y `float()` la convierte a un número decimal, que luego se almacena en la variable `h`.

*   `d = float(input("Ingrese el diámetro de la columna en metros: ").replace(',', '.'))`: Similar a la línea anterior, esta solicita el diámetro de la columna. La parte `.replace(',', '.')` es crucial aquí: reemplaza cualquier coma (`,`) que el usuario pueda haber usado como separador decimal por un punto (`.`) para que `float()` pueda convertir correctamente la cadena a un número decimal. El resultado se guarda en la variable `d`.

*   `# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula los sacos`: Este comentario indica la sección donde se realizan los cálculos.

*   `r = d/2`: Calcula el radio (`r`) de la columna dividiendo el diámetro (`d`) entre 2.

*   `v = 3.1416 * r**2 * h`: Calcula el volumen (`v`) de la columna circular utilizando la fórmula $$V = \pi \cdot r^2 \cdot h$$. Se usa `3.1416` como una aproximación de Pi, `r**2` calcula el radio al cuadrado, y `* h` multiplica por la altura.

*   `sacos = v * 9.5`: Calcula la cantidad total de sacos de cemento necesarios multiplicando el volumen calculado (`v`) por la constante de consumo de cemento (9.5 sacos por metro cúbico).

*   `# SALIDA: Imprimir el volumen y la cantidad de sacos.`: Este comentario indica la sección de salida de los resultados.

*   `print(f"El volumen de la columna es: {v:.2f} m3")`: Imprime el volumen de la columna. `f"...{v:.2f}..."` es una f-string que formatea el valor de `v` para que se muestre con dos decimales.

*   `print(f"La cantidad de sacos necesarios es: {sacos:.2f}")`: Imprime la cantidad de sacos de cemento necesarios, también formateado para mostrar dos decimales.
